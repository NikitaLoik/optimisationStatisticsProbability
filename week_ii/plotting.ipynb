```python
import pylab as plt
```

```python
mySamples = []
myLinear = []
myQuadratic = []
myCubic = []
myExponential = []

for i in range(0, 30):
    mySamples.append(i)
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myExponential.append(1.5**i)
    
```

```python
plt.plot(mySamples, myLinear)
plt.plot(mySamples, myQuadratic)
plt.plot(mySamples, myCubic)
plt.plot(mySamples, myExponential)
plt.show()
```

```python
plt.figure('lin')
plt.xlabel("X")
plt.ylabel("Y")
plt.ylim(0, 1000)
plt.title("linear")
plt.plot(mySamples, myLinear, label = "linear")
plt.legend(loc='upper left')

plt.figure("quad")
plt.clf()
plt.plot(mySamples, myQuadratic)
plt.figure("cube")
plt.clf()
plt.plot(mySamples, myCubic)
plt.figure("exp")
plt.clf()
plt.yscale('log')
plt.plot(mySamples, myExponential)

plt.show()
```
